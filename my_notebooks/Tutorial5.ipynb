```python
import matplotlib.pyplot as plt #displaying images
import PIL #python imaging library - image editing
import urllib #library handling URLs
```

```python
image_url = "https://powerflowsolutions.co.uk/wp-content/uploads/revslider/mountainparallaxheader/mountainpng-1.png"
```

```python
image_data = urllib.request.urlopen(image_url)
image = PIL.Image.open(image_data)
np_image = np.array(image)
```

```python
type(np_image)
```

```python
plt.imshow(np_image)
```

```python
np_image.shape
```

```python
plt.imshow(np_image[200:500,100:500:,3:4])
```

```python
plt.imshow(np_image[200:500,100:500:,0:4])
```

```python
image_red_only = np_image.copy()
image_red_only[:,:,1] = 0 #green channel, shut it down here
image_red_only[:,:,2] = 0 # blue channel, shut it down here
```

```python
plt.imshow(image_red_only)
```
